```python
# You can find covid related data under project's data folder 
base_url = "C:/Users/oztur/deneme-folder/ada-lovelace-academy/data/covid/"
infected_dataset_url = base_url + "time_series_covid19_confirmed_global.csv"
recovered_dataset_url = base_url + "time_series_covid19_recovered_global.csv"
deaths_dataset_url = base_url + "time_series_covid19_deaths_global.csv"
```

#### Solution:

```python
import pandas as pd
import numpy as np
```

```python
# Your code goes here
df_infected = pd.read_csv(infected_dataset_url)
df_recovered = pd.read_csv(recovered_dataset_url)
df_deaths = pd.read_csv(deaths_dataset_url)
```

```python
df_infected.head()
```

```python
df_infected.dtypes
```

```python
grouped_infected = df_infected.groupby('Country/Region').sum()
print(grouped_infected.head())
```

```python
grouped_infected = df_infected.groupby('Country/Region').sum()
grouped_recovered = df_recovered.groupby('Country/Region').sum()
grouped_deaths = df_deaths.groupby('Country/Region').sum()

print("Grouped Infected Data:")
print(grouped_infected.head())

print("\nGrouped Recovered Data:")
print(grouped_recovered.head())

print("\nGrouped Deaths Data:")
print(grouped_deaths.head())
```

```python
# Filter data for specific countries
countries = ['US', 'India', 'Brazil']  
filtered_infected = df_infected[df_infected['Country/Region'].isin(countries)]

# Grouping by Country/Region and finding maximum cases for specific dates
max_cases_specific_countries = filtered_infected.groupby('Country/Region')[['6/9/20', '6/13/20']].max()
print("Max Cases for Specific Dates in Selected Countries:")
print(max_cases_specific_countries)
```

```python
start_date = '5/1/20'
end_date = '8/1/20'

# Filter the data for the specified time interval and 'Country/Region' column
date_range_columns = df_deaths.columns[(df_deaths.columns >= start_date) & (df_deaths.columns <= end_date)]
filtered_deaths = df_deaths[['Country/Region'] + date_range_columns.tolist()]

# Grouping by Country/Region and finding the sum of deaths for the specified time interval
sum_deaths_time_interval = filtered_deaths.groupby('Country/Region').sum()
print(f"Sum of Deaths for {start_date} to {end_date}:")
print(sum_deaths_time_interval.head())
```
